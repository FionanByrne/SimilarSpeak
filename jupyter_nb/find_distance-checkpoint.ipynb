```python
def consonant_dist(word1_list, word2_list):
    dist = 0
    for i, phoneme in enumerate(word1_list, start=0):
        if phoneme in consonants_list:
            dist += phoneme_df.loc[phoneme][word2_list[i]]
    return dist
```

```python
#Return a list of valid phonetic words, obtained by swapping out
def find_similar(word_list, i):
    valid_similar_words = []
    test_list = word_list.copy()
    for con in consonants_list:
        if con != word_list[i]:
            test_list[i] = con
            #ARPAbet uses double whitespace
            test_word = '  '.join(test_list)
            if test_word in valid_phonwords: 
                #print(test_word) 
                valid_similar_words.append(test_word)
    return valid_similar_words
```

```python
consonants_list = [i[0] for i in cmudict.phones() if not i[1] == ['vowel']]
#consonants_list
```

```python
#Create key-value hahmap where unique key = pronounciation, value = word
valid_phonwords = {}
for k, v in cmudict.dict().items():
    key = '  '.join(v[0]) #Convert list -> String
    key = ''.join([i for i in key if not i.isdigit()]) #Remove lexican stress
    #if key not in valid_phonwords.keys(): #Error homophones not included
    valid_phonwords[key] = k
```

Read in input word

```python
num_syllables = 0
while (num_syllables != 1):
    input_word = input()
    input_phonword = pr.phones_for_word(input_word)[0] #Take first result
    num_syllables = pr.syllable_count(input_phonword)
    #num_syllables    
```

```python
#Remove digit indicating phonetic stress
input_phonword = ''.join([i for i in input_phonword if not i.isdigit()])
input_phonword
```

```python
input_phonword_list = input_phonword.split()
phoneme_indexes = list(powerset(range(0,len(input_phonword_list))))
#phoneme_indexes
```

```python
similar_phonwords = []
for index, phoneme in enumerate(input_phonword_list, start=0):
    if phoneme in consonants_list:
        similar_phonwords.extend(find_similar(input_phonword_list, index))
```

```python
similar_words = []
for i in similar_phonwords: similar_words.append(valid_phonwords[i])
similar_words
```

```python
word_rankings = {}
for phonword in similar_phonwords:
    distance = consonant_dist(phonword.split(), input_phonword_list)
    word_rankings[valid_phonwords[phonword]] = distance
word_rankings = sorted(word_rankings.items(), key=operator.itemgetter(1))
word_rankings
```
